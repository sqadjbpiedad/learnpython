**Description:** Add two lists of numbers together effectively.

**Task:** You have the counts of **Males** and **Females** for 5 different barangays. Add them together to get the **Total Population** for each barangay.

* **Males:** `[100, 150, 120, 130, 110]`
* **Females:** `[110, 140, 130, 120, 115]`

```python
np_male = np.array([100, 150, 120, 130, 110])
np_female = np.array([110, 140, 130, 120, 115])

total_population_per_bgy = np_male + np_female
print(f'Total population of each barangay: {total_population_per_bgy}')

# Total population of all barangays
total_population = np.sum(total_population_per_bgy)
print(f'Total population of each barangay: {total_population}')

# Average population male and female
ave_male = np.mean(np_male)
print(f'Average population of male: {ave_male}')
ave_female = np.mean(np_female)
print(f'Average population of female: {ave_female}')

# Difference of barangay population from the average
```

## Exercise 3: Finding the Highest Inflation Rate

## Exercise 5: Filter Large Households

**Description:** Find numbers that are bigger than a specific value.

**Task:** You have a list of household family sizes. Find and print only the family sizes that have **more than 5** members.

* **Family Sizes:** `[3, 4, 8, 2, 6, 10, 4]`

## Long Exercise: Annual Report on Violence Against Women (VAW)

**Scenario:**
You are preparing a report on the **Monthly Reported Cases of Violence Against Women (VAW)** in a specific municipality for the year 2024.

**Your Tasks:**

1.  **Create the Data:** Convert the list of 12 months into a NumPy array.
2.  **Total Annual Cases:** Calculate the total number of VAW cases reported for the whole year.
3.  **Average Monthly Cases:** Calculate the average number of cases reported per month.
4.  **Peak Incident Month:** Identify the highest number of cases recorded in a single month.
5.  **Identify High-Risk Months:** Find and print all months where the reported cases were **greater than 20** (indicating a surge).

**Raw Data (Jan-Dec):**
`15, 18, 12, 14, 25, 28, 10, 8, 30, 22, 15, 19`

```python
data = [
    {"month": "January", "case": 15},
    {"month": "February", "case": 18},
    {"month": "March", "case": 12},
    {"month": "April", "case": 14},
    {"month": "May", "case": 25},
    {"month": "June", "case": 28},
    {"month": "July", "case": 10},
    {"month": "August", "case": 8},
    {"month": "September", "case": 30},
    {"month": "October", "case": 22},
    {"month": "November", "case": 15},
    {"month": "December", "case": 19}
]

months = np.array([month['month'] for month in data])
incidents = np.array([incident['case'] for incident in data])

total_annual_case = np.sum(incidents)
print(f'Total Annual Cases: {total_annual_case}')

average_monthly_cases = np.mean(incidents)
print(f'Average Monthly Cases: {average_monthly_cases}')

peak_incident_month = np.max(incidents)
print(f'Peak Incident Month: {peak_incident_month}')

high_risk_months = months[incidents > 20]
print(f'High Risk Months: {high_risk_months}')
```
