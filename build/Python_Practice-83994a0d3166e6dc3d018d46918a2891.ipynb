```python
class Farm:
    def __init__(self, name, crop, hectares):
        # Attributes: Data specific to this farm
        self.name = name
        self.crop = crop
        self.hectares = hectares

    # Method: An action the farm can perform
    def describe_farm(self):
        return f"{self.name} plants {self.crop} on {self.hectares} hectares."

# Creating an Object (Instance)
my_farm = Farm("Maligaya Estate", "Palay", 10)

# Calling a Method
print(my_farm.describe_farm())

print(my_farm)
```

```python
class Farm:
    def __init__(self, name, crop, hectares):
        self.name = name
        self.crop = crop
        self.hectares = hectares

    def estimate_production(self):
        """Assume standard yield of 4.5 tons/ha"""
        return self.hectares * 4.5

    # OVERRIDING the standard print function
    def __str__(self):
        return f"FARM: {self.name} | CROP: {self.crop} | AREA: {self.hectares} ha"

farm_1 = Farm("Maligaya Fields", "Corn", 5)

print(farm_1) 
print(f"Estimated Harvest: {farm_1.estimate_production()} tons")
```

```python
class Livestock:
    """
    A class to represent livestock.

    Attributes:
        species (str): The species of the livestock.
        count (int): The number of animals.
    """

    def __init__(self, species: str, count: int):
        """
        Initialize a Livestock object.

        Args:
            species (str): The species of the livestock.
            count (int): The number of animals.
        """
        self.species = species
        self.count = count

    def check_status(self) -> str:
        """
        Determine the status of the herd based on its size.

        Returns:
            str: "Small Herd" if count < 5, otherwise "Commercial Herd".
        """
        if self.count < 20:
            return "Small Herd"
        else:
            return "Commercial Herd"


# Example usage:

my_animals = Livestock("Carabao", 50)
print(my_animals.check_status())

my_animals = Livestock("Cow", 8)
print(my_animals.check_status())
```

```python
tank_level = 0
max_capacity = 50
pump_speed = 3
halfway = 0

while tank_level < max_capacity:
    tank_level += pump_speed
    if (tank_level >= max_capacity/2) & (halfway == 0):
        print("Halfway there!")
        halfway = 1
    print(f'Tank level: {tank_level}')
    if tank_level == max_capacity:
        print("Tank is full.")
    elif tank_level > max_capacity:
        print(f'Overflow by {tank_level - max_capacity}')
```

```python
# Example: Calculating Yield per Hectare (with risky math)
def calculate_yield(production, hectares):
    try:
        result = production / hectares
        return f"{result:.2f} tons/ha"
    except ZeroDivisionError:
        return "Error: Cannot divide by zero (Hectares is 0)"
    except TypeError:
        return "Error: Inputs must be numbers"

# Test cases
print(calculate_yield(100, 10))  # Normal
print(calculate_yield(100, 0))   # Error: Division by Zero
print(calculate_yield("100", 5)) # Error: Wrong Type
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data: Palay Production by Region (in Metric Tons)
data = {
    'Region': ['Central Luzon', 'Cagayan Valley', 'Western Visayas', 'Bicol', 'Ilocos'],
    'Production_MT': [350000, 280000, 210000, 150000, 180000]
}
df_crops = pd.DataFrame(data)

# Sort the data
df_sorted = df_crops.sort_values('Production_MT', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
# orient='h' makes it horizontal
sns.barplot(x='Region', y='Production_MT', data=df_sorted, color='teal')

# Polish
plt.title("Central Luzon leads in Palay Production", loc='left', fontsize=14, fontweight='bold')
plt.ylabel("Metric Tons")
plt.show()
```

```python
import numpy as np

# Simulating Farm Size Data (Right-Skewed)
# Most farms are small (1-3 ha), few are large
np.random.seed(42)
farm_sizes = np.random.exponential(scale=2.0, size=1000)

plt.figure(figsize=(10, 6))

# KDE=True draws the smooth curve showing the skew
sns.histplot(farm_sizes, kde=True, color='brown', element="step")

# Add a reference line for the mean
avg_size = np.mean(farm_sizes)
plt.axvline(avg_size, color='red', linestyle='-.', label=f'Mean: {avg_size:.5f} ha')

plt.title("Distribution of Farm Sizes (Most < 3 ha)", loc='left')
plt.xlabel("Hectares")
plt.legend()
plt.show()
```
